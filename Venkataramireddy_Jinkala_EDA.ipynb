```python
import pandas as pd

# Load datasets
customers_df = pd.read_csv('C:/Users/Jinkala Venkat/Downloads/Customers.csv')
products_df = pd.read_csv('C:/Users/Jinkala Venkat/Downloads/Products.csv')
transactions_df = pd.read_csv('C:/Users/Jinkala Venkat/Downloads/Transactions.csv')

# Convert date columns
customers_df['SignupDate'] = pd.to_datetime(customers_df['SignupDate'])
transactions_df['TransactionDate'] = pd.to_datetime(transactions_df['TransactionDate'])

# Merge datasets
merged_df = transactions_df.merge(customers_df, on='CustomerID', how='left').merge(products_df, on='ProductID', how='left')

# Descriptive statistics
summary_stats = merged_df.describe()

# Transactions by region
transactions_by_region = merged_df['Region'].value_counts()

# Top categories by quantity
top_categories = merged_df.groupby('Category')['Quantity'].sum().sort_values(ascending=False)

# Monthly revenue trends
merged_df['Month'] = merged_df['TransactionDate'].dt.to_period('M')
monthly_revenue = merged_df.groupby('Month')['TotalValue'].sum()

# Print results
print("Summary Statistics:")
print(summary_stats)
print("\nTransactions by Region:")
print(transactions_by_region)
print("\nTop-Selling Categories:")
print(top_categories)
print("\nMonthly Revenue Trends:")
print(monthly_revenue)
```
